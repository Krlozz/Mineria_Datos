```python
import pandas as pd
import apriori_python as apr
import math
```

```python
dataset = pd.read_excel('C:/Users/Krlozz/Downloads/Dataset Basket.xls')
```

```python
print(dataset.head)
```

```python
dataset = dataset.loc[:, ['confectionery', 'fruitveg', 'freshmeat', 'dairycannedveg', 'cannedmeal', 'frozenmeal', 'beerwine', 'softdrink', 'fish', 'bread', 'Soap']]
```

```python
dataset.replace({'F': 0, 'T': 1}, inplace=True)
```

```python
print(dataset.head)
```

```python
datos1 = []
```

```python
for i in range(len(dataset)):
    datos1.append([str(int(dataset.values[i, j])) for j in range(0,dataset.shape[1]) if not math.isnan(dataset.values[i, j])])
```

```python
print(datos1)
```

```python
minsup = 0.1
minconf = 0.4

pf, reglas = apr.apriori(datos1, minSup = minsup, minConf = minconf)
```
